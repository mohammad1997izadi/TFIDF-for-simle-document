```python
import math
from collections import OrderedDict

data = {
        1:'درس بازیابی اطلاعات در مقطع تحصیلات تکمیلی ارائه می شود',
        2:'درس بازیابی اطلاعات در مقطع کارشناسی ارائه نمی شود',
        3:'درس داده کاوی در مقطع کارشناسی ارشد ارائه می شود',
        4:'درس داده کاوی در مقطع کارشناسی ارائه می شود',
        5:'انتخاب ریاست جمهوری آمریکا درآبان ماه برگزار می شود ',
        6:'دوره اول ریاست جمهوری آقای رییسی شروع شده است',
        7:'وضعیت محیط زیست ایران جالب نیست',
        8:'تیم پرسپولیس اخیرا در چند فینال لیگ قهرمانان آسیا شرکت کرده است',
        9:'دانشگاهها به صورت مجازی در دوران کرونا فعالیت می کنند',
        10:'این اولین تمرین درس بازیابی پیشرفته اطلاعات می باشد'
        }

def ranking(query, doc_dict, words_bag):
    
        
    
    query_vocab = []
    for word in query.split(' '):
        if word not in query_vocab:
            query_vocab.append(word)

            
    tf_idf_scr = {}
    for doc_id in doc_dict.keys():
        tf_idf_scr[doc_id] = {}
    for word in query_vocab:
        for doc_id,doc in doc_dict.items():
            tf = TF(word, words_bag)
            df, idf = DF_and_IDF(word, words_bag, doc_dict)
            tfidf = tf_idf(tf, idf, doc_dict)

            tf_idf_scr[doc_id][word] = tfidf[doc_id]
    

    
    query_re = {}
    for word in query_vocab:`
        query_re[word] = query.lower().split(' ').count(word)
    
    
    relevance_scores = {}
    for doc_id in doc_dict.keys():
        score = 0
        for word in query_vocab:
            score += query_re[word] * tf_idf_scr[doc_id][word]
        relevance_scores[doc_id] = score
    sorted_value = OrderedDict(sorted(relevance_scores.items(), key=lambda x: x[1], reverse = True))
    top_10 = {k: sorted_value[k] for k in list(sorted_value)[:10]}
    return top_10
            
            
            
            
            

def TF(word, words_bag):
    
    result = {}
    for key in range(1, 11):
        counter = 0
        size = len(words_bag[key])
        for w in words_bag[key]:
            if w == word:
                counter+=1
                
        result[key] = counter / size   
    return result


def DF_and_IDF(word, words_bag, data):
    
    total_doc = len(data)
    df = 0
    for key in range(1, total_doc+1):
        for w in words_bag[key]:
            if w == word:
                df += 1
                break
            
    if df == 0:
        idf = 0
    else:
        idf = math.log(total_doc / df)
    
    return df, idf
    
    
def tf_idf(tf, idf, data):
    total_doc = len(data)
    
    tfidf = {}
    for key in range(1, total_doc+1):
        tfidf[key] = tf[key]*idf
    
    
    return tfidf

words_bag = {}

for key, value in data.items():
    words_bag[key] = value.split(' ')
    
    
word = input('Enter a word for: ');
tf = TF(word, words_bag)
df, idf = DF_and_IDF(word, words_bag, data)
tfidf = tf_idf(tf, idf, data)

print('tf: ')
print(tf)
print('df, idf :')
print(df, idf)
print('tf_idf: ')
print(tfidf)


query = input('Enter a query for ranking: ');
print(ranking(query, data, words_bag))
```
